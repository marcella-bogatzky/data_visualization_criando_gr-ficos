# **Aula 1 - Conhecendo a biblioteca Matplotlib**


##**Conhecendo os dados**

### Importando os dados

Contexto: Uma empresa que presta serviços de consultoria e aconselhamento para pessoas que querem migrar do Brasil para o Canadá deseja melhorar os seus serviços. Para isso, temos alguns arquivos que foram disponibilizados para fazer uma análise das tendências de imigração do Brasil dos últimos anos.

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

###Realizando alguns tratamentos

**Primeira tarefa:** analisar as tendências de imigração do Brasil ao longo dos anos.

```python
df.set_index('País', inplace=True) #Setar o índice com base nos países
```

```python
anos = list(map(str, range(1980, 2014))) #Armazenar intervalo de tempo em uma variável
#list - cria lista
#map - aplica no intervalo a informação passada
#str - informação passada em map (transforma informação em string)
#range - intervalo (não considera o último valor)
```

```python
anos
```

### Extraindo a série de dados para o Brasil

```python
brasil = df.loc['Brasil', anos] #Selecionar apenas os anos do Brasil
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()} #Armazenar em um dicionário
#'ano' e 'imigrantes' são as chaves do dicionario
#index - vai puxar os dados da primeira coluna da variável brasil
#value - puxa os dados da segunda coluna da variável brasil
#tolist - função que faz a conversão em uma lista
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

##**Criando o primeiro gráfico com Matplotlib**

> Documentação da biblioteca: https://matplotlib.org/

**Segunda tarefa:** criar um gráfico para entender as tendências de imigração do Brasil.

Para isso, vamos utilizar biblioteca visualização de dados em Pyhton: a biblioteca `Matplotlib`. Para criar gráficos, vamos utilizar o `módulo pyplot` específico dessa biblioteca.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

- Escolher quais anos aparecerem no eixo X com `plt.xticks`
- Alterar os ticks do eixo Y com `plt.yticks`
- Exibir gráfico com `plt.show()`
- Modificar o tamanho do gráfico com a função `plt.figure()`
- Adicionar título com `plt.title()`
- Adicionar rótulos aos eixos X e Y usando `plt.xlabel()` e `plt.ylabel()`


```python
plt.figure(figsize=(8,4)) #Define o tamanho do gráfico em polegadas
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #Primeiro colocamos os dados que serão analizados horizontalmente, depois os que serão analisados verticalmente
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) #É possível realizar o mesmo no eixo y utilizando yticks
plt.show() #Faz com que seja exibido somente o gráfico (precisa ser sempre o último item)
```

#**Aula 2 - Criando figuras com Matplotlib**


##**Criando uma figura**

Terceira tarefa: criar uma figura.

Para isso, vamos utilizar a função `plt.subplots`.

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.show()
```

- Ajustar a frequência dos rótulos do eixo X
- Adicionar título
- Adicionar rótulos

```python
fig, ax = plt.subplots(figsize=(8,4)) #A função subplot deve ser atribuída a outras variáveis
#Fig - permite a criação de uma figura (em branco)
#Ax (axes) - representa os eixos
#figsize - tamanho da figura
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano') #Rótulo do eixo X
ax.set_ylabel('Número de imigrantes') #Rótulo do eixo Y
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#xaxis representa o eixo x
#set_major_locator são os dados do eixo x
plt.show()
```

##**Criando subplots em uma direção**



Enquanto o gráfico de linhas é interessante para mostrar tendências ao longo do tempo, podemos também utilizar o boxplot (gráfico de caixa), que é um gráfico para mostrar algumas estatísticas em relação aos nossos dados.

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
#subplot represeta quantos gráficos (em linha e coluna) serão adicionados em uma figura

#Primeiro subplot
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid() #Adiciona uma grade
#axs[0] diz que todas essas informações fazem parte do gráfico índice 0

#Segundo subplot
axs[1].boxplot(dados_brasil['imigrantes']) #Em boxplot passamos apenas os dados do eixo Y
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')

plt.show()
```

### Adicionando grid para entender melhor os valores

```python
# Criando a figura
fig, axs = plt.subplots(1, 2, figsize=(15,5))

# Subplot 1: Gráfico de linhas
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

# Subplot 2: Boxplot
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Box-plot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid() #Adiciona uma grade

plt.show()
```

**Boxplot**

No boxplot, temos uma "caixa" com uma linha inferior e outra superior. Essas extremidades representam a amplitude, ou seja, o valor mínimo na parte inferior e o valor máximo na parte superior. Isso significa que o valor mínimo é de aproximadamente 100 imigrantes (vamos confirmar isso em breve) e o valor máximo está entre 1.500 e 2.000.

Agora, vamos entender a "caixa" do boxplot. Ela possui três partes: a primeira parte inferior indica que, em 25% dos anos, o número de imigrantes foi cerca de 400 a 500. A linha central, em laranja, indica que em 50% desse período, o número de imigrantes foi entre 500 e 1.000 (aproximadamente 700). A linha superior mostra que em 75% dos anos, o número de imigrantes foi de cerca 1.000.

##**Criando subplots em duas direções**

```python
# Criando a figura
fig, axs = plt.subplots(2, 2, figsize=(10, 6)) #Lembrando que na primeira parte do subplot, definimos quantas linhas e colunas terá nossa figura. Nesse caso, 2 de cada

# Subplots
#Brasil
axs[0, 0].plot(df.loc['Brasil', anos]) #O axs[] define a localização dentro do figura. Nesse caso, ficará na primeira linha e primeira coluna (índice sempre inicia em zero)
axs[0, 0].set_title('Brasil')

#Colômbia
axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

#Argentina
axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

#Peru
axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

plt.show()
```

- Ajustar o espaçamento
- Alterar a frequência dos ticks do eixo X em todos os subplots
- Adicionar rótulos para os eixos X e Y
- Adicionar um título geral

```python
# Criando a figura
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Ajustando o espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.3) #hspace é o espaçamento horizontal | wspace é o espaçamento vertical

# Adicionando um título geral para os subplots
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')

# Subplots
axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

# Alterando a frequência dos ticks do eixo X em todos os subplots
for ax in axs.flat: #for é um comando de repetição | .flat faz com que o loop percorra todos os subplots da figura
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Adicionando rótulos para os eixos X e Y
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')

plt.show()
```

### Definindo a mesma escala no eixo Y em todos os subplots



```python
# Criando a figura
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Ajustando o espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Adicionando um título geral para os subplots
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')

# Subplots
axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

# Alterando a frequência dos ticks do eixo X em todos os subplots
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Adicionando rótulos para os eixos X e Y
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')

##### Definindo a mesma escala no eixo Y em todos os subplots
ymin = 0
ymax = 7000
for ax in axs.ravel(): #método .ravel transforma um array bidimensional em um array unidimensional
    ax.set_ylim(ymin, ymax)

plt.show()
```

# **Aula 3 - Customizando com Matplotlib**

##**Personalizando visualizações com Matplotlib**

- Alterar o tamanho do título
- Alterar a posição do título
- Alterar o tamanho dos ticks
- Alterar a espessura da linha
- Adicionar marcadores

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, marker='o') #lw altera a expessura da linha do gráfico | marker adiciona marcadores na linha (a letra o adiciona uma bolinha)

ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left') #fontsize altera o tamanho da fonte | loc altera a posição
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.set_xlabel('Ano', fontsize=14)

ax.yaxis.set_tick_params(labelsize=12) #labelsize altera o tamanho da fonte dos "ticks" (informações do eixo)
ax.xaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
```

### Adicionando grid e mudando o estilo do grid

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.set_xlabel('Ano', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.grid(linestyle = '--') #plt.grid adiciona a grade | linestyle define o estilo da grade
plt.show()
```

##**Alterando as cores**

> As cores podem ser encontradas em: https://matplotlib.org/stable/gallery/color/named_colors.html

Algumas opções de cores incluem:

- *Base colors*;
- *Tableau palette*;
- CSS *colors*.

As ***base colors*** são as cores mais básicas pré-definidas e na frente do retângulo de cada cor temos a letra que representa a cor. São elas azul(b), verde(g), vermelho(r), ciano(c), magenta(m), amarelo(y), preto(k) e branco(w).

A ***tableau pallet*** contém outras cores pré-definidas e bem distintas entre si, sendo ideal para trabalhar com dados categóricos. Por fim, temos as **CSS *colors***, usadas para aplicações web e criação de interfaces.

### Mudando apenas uma cor

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g') #color define a cor, "g" é a base color verde

ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.set_xlabel('Ano', fontsize=14)

ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
```

### Mudando várias cores

Como alterar as cores quando há múltiplas variáveis?

Para ilustrar isso, vamos comparar o Brasil com todos os países da América do Sul. Inclusive avaliaremos os números totais de imigrantes para cada um desses países para compararmos como o Brasil está posicionado em relação a eles durante esse período.

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
# Lista de cores
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

# Criando a figura
fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(america_sul.index, america_sul['Total'], color=cores)
#ax.bar cria o gráfico em barras
#primeiro é definido o eixo x (america_sul.index)
#depois é definido o eixo y (america_sul['Total'])

ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.set_xlabel('')

ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

### Deixando o gráfico na horizontal

```python
# Lista de cores
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

# Criando a figura
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)
#ax.barh cria o gráfico em barras na horizontal
#primeiro é definido o eixo y (america_sul.index)
#depois é definido o eixo x (america_sul['Total'])

ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')

ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

### Deixando as barras do maior para o menor valor

```python
america_sul_sorted = america_sul.sort_values(by='Total', ascending=True)
#sort_values ordena os valores
#by define qual dado será ordenado
#ascending define a ordenação do maior para o menor
```

```python
america_sul_sorted
```

```python
# Lista de cores
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

# Criando a figura
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

### Destacacando informações

```python
# Criando uma lista de cores para destacar o Brasil
cores = []
for pais in america_sul_sorted.index: #repetição feita para atribuir a cor verde a barra do país quando for o Brasil e a cor cinza quando não for
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

# Criando a figura
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

# Loop para adicionar o valor máximo de cada barra
for i, v in enumerate(america_sul_sorted['Total']): #função enumerate que vai iterar sobre cada item na coluna do valor total
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center') #ax.text adiciona texto no eixo x
    #v + 20 indica onde o texto vai ser correspondente na barra
    #i vai ser a posição y do texto
    #str(v) indica que vamos converter o tamanho da barra em uma string
    #ha indica o alinhamento horizontal do texto
    #va indica o alinhamento vertical do texto

plt.show()
```

### Retirando frame, o eixo x e os ticks ao lado do nome dos países:


```python
# Criando uma lista de cores para destacar o Brasil
cores = []
for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

# Criando a figura
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

# Loop para adicionar o valor máximo de cada barra
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

# Retirando o Frame (retângulo que encobre os gráficos)
ax.set_frame_on(False)

# Retirando o eixo X
ax.get_xaxis().set_visible(False)

# Removendo todos os ticks dos eixos X e Y
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```
